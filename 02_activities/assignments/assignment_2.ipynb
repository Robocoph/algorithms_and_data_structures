  # Question Two: Path to Leaves

  Given the `root` of a binary tree, return all root to leaf paths in any order.

  ## Examples

  ### Example 1

  ![](./images/q1_ex1.png)

  Input: `root = [1, 2, 2, 3, 5, 6, 7]` *What traversal method is this?*

  Output: [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]

  ### Example 2

  ![](./images/q1_ex3.png)

  Input: `root = [10, 9, 7, 8]`

  Output: [[10, 7], [10, 9, 8]]

</details>

#### Starter Code for Question 2

```python
# This looks like DFS implementation

# Definition for a binary tree node.
class TreeNode(object):
  def __init__(self, val = 0, left = None, right = None):
    self.val = val # node value
    self.left = left # left child
    self.right = right # right child

def bt_path(root): 
    result = [] # This will be storing all the paths from the roots to leaves

    def dfs(node,path): # this is a helper function for DFS implementation
        
      # Check for node
        if not node: # If we reach an empty child node then no use proceeding
           return
        
        path.append(node.val) # add the current node to the path

      # Check for leaf node
        if not node.left and not node.right: # no children on left and right
           result.append(list(path)) # Save the path to the list
        else:
           # Recursion for any subtrees on left and right
           dfs(node.left, path)
           dfs(node.right, path)
      # Remove current node from the path
        path.pop()

    dfs(root,[])

    return result
```

```python
# Testing the function

# Creating the nodes manually using Example 1
n1 = TreeNode(1)
n2_left = TreeNode(2)
n2_right = TreeNode(2)
n3 = TreeNode(3)
n5 = TreeNode(5)
n6 = TreeNode(6)
n7 = TreeNode(7)

# Link nodes to form the tree
n1.left = n2_left
n1.right = n2_right
n2_left.left = n3
n2_left.right = n5
n2_right.left = n6
n2_right.right = n7

# n1 is the root
paths = bt_path(n1)

print(paths)
```

```python
# Example: 
# Input: [9, 3, 2, 5, 2, 3]
# Output: 2


# Walkthrough of Partner’s Example: [1, 2, 3, 1, 4, 4]

# Step 1: Read 1, add it to the set 
# Step 2: Read 2, add it to the set
# Step 3: Read 3, add it to the set
# Step 4: Read 1, already in set, return the value 1 from set
```


-   Copy the solution your partner wrote. 


```python
# Partner's solution 

from typing import List

def first_duplicate(nums: List[int]) -> int:

    X = set()  # Hash set to store seen values
    
    for num in nums:
        if num in X:
            return num  # Found first duplicate
        X.add(num)
    
    return -1  # No duplicates found
```


-   Explain why their solution works in your own words.


```python
# This code functions by keeping track of all numbers it has seen using a set, which allows fast lookup (constant time). 

# As it loops through the list:
# If a number is already in the set, it means we’ve seen it before then return it.
# If not, we add it to the set and continue.
# If we go through the whole list and find no repeats, then we return -1.
# This way we can guarantee that the first duplicate is returned as soon as it is found.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Time Complexity: O(n), where n is the number of items in the list. 
#   We go through the list only once and each lookup/add operation in a set is very fast (O(1)).

# Space Complexity: O(n) in the worst case. 
#   If there are no duplicates, we end up storing every number in the set.
```


-   Critique your partner's solution, including explanation, and if there is anything that should be adjusted.


```python
# Strenghts:
# Solution is well-written, clean and executes efficiently.
# Duplicate detection appropriately with set.
# Checks duplicates at the first instance and solves the problem.

# Improvement Opportunity:
# It might be a good production practice if we check for empty first. Once empty list encountered:
#   We immediately return -1 and skip creation of set and entering the loop. 
#   It will be a good defensive practice.
# Would have been great if the variable was named more descriptively. For example,
# Instead of X, it could have been read = set() or seen = set()
# Although not covered in this course, some docstring documentation would have been great.
```
