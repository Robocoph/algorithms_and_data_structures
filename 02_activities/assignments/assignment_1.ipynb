## Question Two: Valid Bracket Sequence
**Description**  
Given a string containing only the characters `'('`, `')'`, `'{'`, `'}'`, `'['`, and `']'`, determine if the input string is a **valid bracket sequence**.  
A string is valid if:
- Open brackets are closed by the same type of brackets, and
- Open brackets are closed in the correct order.

**Examples**
```python
Input: s = "([]{})"
Output: True
```
```python
Input: s = "([)]"
Output: False
```
```python
Input: s = "()[]{}"
Output: True
```
```python
Input: s = "[{]}"
Output: False
```

**Question 2 Starter Code**

````python
def is_valid_brackets(s: str) -> bool:
    # We will initiate an empty stack to match each opening with a closing bracket.
    bracket_stack = []
    # We will also generate a mapping dictionary
    bracket_map = {')' : '(',
                   '}' : '{',
                   ']' : '['}
    
    for char in s: 
        if char in bracket_map.values():  # .values() contains all the opening bracket values
            bracket_stack.append(char) # if the character/string is an opening bracket then append to the stack

        elif char in bracket_map: # Now checking for closing brackets 
            if not bracket_stack or bracket_stack[-1] != bracket_map[char]: # if the closing bracket not matched with an opening bracket then it should print false
                return False
            bracket_stack.pop() # opening and closed backet matched, so we are poppig the stack by removing the opening bracket

        else:
            return False 
    
    return len(bracket_stack) == 0

````

````python
def is_valid_brackets(s: str) -> bool:
    # We will initiate an empty stack to match each opening with a closing bracket.
    bracket_stack = []
    # We will also generate a mapping dictionary
    bracket_map = {')' : '(',
                   '}' : '{',
                   ']' : '['}
    
    for char in s: 
        if char in bracket_map.values():  # .values() contains all the opening bracket values
            bracket_stack.append(char) # if the character/string is an opening bracket then append to the stack

        elif char in bracket_map: # Now checking for closing brackets 
            if not bracket_stack or bracket_stack[-1] != bracket_map[char]: # if the closing bracket not matched with an opening bracket then it should print false
                return False
            bracket_stack.pop() # opening and closed backet matched, so we are poppig the stack by removing the opening bracket

        else:
            return False 
    
    return len(bracket_stack) == 0

# Lets run Input 1

# Input 1 
s = "[({})]"

# Execute the function
is_valid_brackets(s)
````


-   Explain why your solution works


````python
# Explanation: 

# The solution adopts a stack solution. Each time we encounter an opening bracket, the bracket is appended to the stack
# For each closing bracket encountered, we check if the top of the stack has the opening bracket. 
# If we encounter a matching opening bracket, we pop it.
# If we encounter an unmatching opening bracekt, then we throw False.
# If the stack is empty but it must not be then the sequence is invalid too.
# In the end if the stack is empty after all the matching then all brackets were properly matched.
````


-   Explain the problem’s time and space complexity


````python
# Time Complexity: We need to iterate through the string once to determine if all brackets are closed appropriately. We also only need to read a character only once. Hence it is O(1) for each character, and O(n) is n characters.

# Space Complexity: In this example, we are using a stack to store the opening brackets. The worst case scenario is if the input is all open brackets like s = "(((((([[[[[{{{{{". 
# In this worst case we must store all the opening brackets into the stack and nothing will be popped. This will lead ot stack growing to a size of n inputs.

# Hence, worst case space complexity is O(n)
````
